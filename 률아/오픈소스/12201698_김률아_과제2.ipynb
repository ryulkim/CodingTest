```python
import re
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

url="./Downloads/ml-100k/ml-100k/" #데이터가 들어있는 경로

path_data=url+"u.data"
path_item=url+"u.item"
path_occupation=url+"u.occupation"
path_user=url+"u.user"

data=[]
genre=['unknown','Action','Adventure','Animation','Children\'s','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']
item=[]
occupation=[]
user=[]

f_item=open(path_item,mode="r",encoding="ISO-8859-1")
f_data=open(path_data,mode="r",encoding="utf-8")
f_user=open(path_user,mode="r",encoding="utf-8")
f_occupation=open(path_occupation,mode="r",encoding="utf-8")

line=f_data.readline()
while line:
    line=line.rstrip("\n")
    line=line.split('\t')
    data.append(line)
    line=f_data.readline()


line=f_item.readline()
while line:
    line=line.rstrip("\n")
    line=re.split('[|||]',line)
    item.append(line)
    line=f_item.readline()
    
line=f_user.readline()
while line:
    line=line.rstrip("\n")
    line=line.split('|')
    user.append(line)
    line=f_user.readline()
    
line=f_occupation.readline()
while line:
    line=line.rstrip("\n")
    occupation.append(line)
    line=f_occupation.readline()
    

    
f_data.close()
f_item.close()
f_user.close()
f_occupation.close()

df_data=pd.DataFrame(np.array(data),columns=['user id','item id','rating','timestamp'])
df_item=pd.DataFrame(np.array(item),columns=['movie id', 'movie title','release date','video release date','IMDb URL','unknown','Action','Adventure','Animation','Children\'s','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western'])
df_user=pd.DataFrame(np.array(user),columns=['user id','age','gender','occupation','zip code'])
df_data['rating']=pd.to_numeric(df_data['rating'])
```

```python
#2
#A
#모든 유저에서 평점(rating)의 개수가 많은 100개의 영화
list=df_data['item id'].value_counts()[:100]
s_to_df=list.reset_index()
s_to_df.columns=['movie id','count']
df_movie=df_item.loc[df_item['movie id'].isin(list.index),:'movie title']
df_movie=pd.merge(left=df_movie,right=s_to_df,how="inner",on="movie id")
df_movie.sort_values(by='count', ascending=False)
```

```python
#B
#모든 유저에서 장르별 평점의 평균
list=df_data[['item id','rating']]

df_movie_id=df_data.copy()
df_movie_id['sum']=list.groupby(['item id']).transform('sum')
df_movie_id['count']=list.groupby(['item id']).transform('count')
df_movie_id=df_movie_id[['item id', 'count','sum']]
df_movie_id=df_movie_id.rename(columns={'item id':'movie id'})
df_movie_id=df_movie_id.drop_duplicates(['movie id'],keep='first')

df_movie=df_item.loc[df_item['movie id'].isin(df_movie_id['movie id'])]
df_movie=pd.merge(left=df_movie,right=df_movie_id,how="inner",on="movie id")
df_movie=df_movie.loc[:,'unknown':'sum']

genre_mean=[]

for i in range(0,19):
    list=df_movie[df_movie.iloc[:,i]=='1']
    rating=list['sum'].sum()/list['count'].sum()
    genre_mean.append([genre[i],rating])
df_genre_mean=pd.DataFrame(np.array(genre_mean),columns=['genre','rating'])
df_genre_mean['rating']=pd.to_numeric(df_genre_mean['rating'],errors='coerce').fillna(0, downcast='infer')
df_genre_mean.sort_values(by='rating', ascending=False)
```
